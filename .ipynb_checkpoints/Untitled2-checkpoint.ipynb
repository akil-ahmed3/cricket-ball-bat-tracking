```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import imutils
import os
import re
```

```python
video = "./videos/2.mp4"

cap = cv2.VideoCapture(video)
cnt = 0

if (cap.isOpened() == False):
    print("File can not be opened")
```

```python
ret, first_frame = cap.read()

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret == True:
        roi = frame[:800,:]
        thresh=600
        end = roi.shape[1] - thresh
        roi = roi[:,thresh:end]
        
        cv2.imshow("image",roi)
        
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
            
        cv2.imwrite('./frames/'+str(cnt)+'.png',roi)
        cnt=cnt+1
    else: 
        break

cv2.destroyAllWindows()    
```

```python
frames = os.listdir('frames/')
frames.sort(key=lambda f: int(re.sub('\D', '', f)))

#reading frames
images=[]
for i in frames:
    img = cv2.imread('frames/'+i)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.GaussianBlur(img,(11,11),0)
    images.append(img)

images=np.array(images)
```

```python
nonzero=[]
for i in range((len(images)-1)):
    
    mask = cv2.absdiff(images[i],images[i+1])
    mask = cv2.adaptiveThreshold(mask, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
    num = np.count_nonzero((mask.ravel()))
    nonzero.append(num)
    
    
x = np.arange(0,len(images)-1)
y = nonzero

plt.figure(figsize=(20,4))
plt.scatter(x,y)
```

```python
import pandas as pd
folders=os.listdir('images/')

images=[]
labels= []
for folder in folders:
    files=os.listdir('images/'+folder)
    for file in files:
        img=cv2.imread('images/'+folder+'/'+file,0)
        img=cv2.resize(img,(25,25))
        
        images.append(img)
        labels.append(folder)

images = np.array(images)
features = images.reshape(len(images),-1)
```

```python
for i, label in enumerate(labels):
    if label == "miscellaneous":
        labels[i] = 1
    else:
        labels[i] = 0
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_val,y_tr,y_val = train_test_split(features,labels, test_size=0.2, stratify=labels,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier(max_depth=3) 
rfc.fit(x_tr,y_tr)
```

```python
from sklearn.metrics import classification_report
y_pred = rfc.predict(x_val)
print(classification_report(y_val,y_pred))
```
